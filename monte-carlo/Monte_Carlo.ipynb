### Part 1: MC Prediction

In this section, you will write your own implementation of MC prediction (for estimating the action-value function).  

We will begin by investigating a policy where the player _almost_ always sticks if the sum of her cards exceeds 18.  In particular, she selects action `STICK` with 80% probability if the sum is greater than 18; and, if the sum is 18 or below, she selects action `HIT` with 80% probability.  The function `generate_episode_from_limit_stochastic` samples an episode using this policy. 

The function accepts as **input**:
- `bj_env`: This is an instance of OpenAI Gym's Blackjack environment.

It returns as **output**:
- `episode`: This is a list of (state, action, reward) tuples (of tuples) and corresponds to $(S_0, A_0, R_1, \ldots, S_{T-1}, A_{T-1}, R_{T})$, where $T$ is the final time step.  In particular, `episode[i]` returns $(S_i, A_i, R_{i+1})$, and `episode[i][0]`, `episode[i][1]`, and `episode[i][2]` return $S_i$, $A_i$, and $R_{i+1}$, respectively.

````python
def generate_episode_from_limit_stochastic(bj_env):
    episode = []
    state = bj_env.reset()
    while True:
        probs = [0.8, 0.2] if state[0] > 18 else [0.2, 0.8]
        action = np.random.choice(np.arange(2), p=probs)
        next_state, reward, done, info = bj_env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode
````

Now, you are ready to write your own implementation of MC prediction.  Feel free to implement either first-visit or every-visit MC prediction; in the case of the Blackjack environment, the techniques are equivalent.

Your algorithm has three arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `generate_episode`: This is a function that returns an episode of interaction.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

````python
def mc_prediction_q(env, num_episodes, generate_episode, gamma=1.0, mc_approx='first'):
    # initialize empty dictionaries of arrays
    returns_sum = defaultdict(lambda: np.zeros(env.action_space.n))
    N = defaultdict(lambda: np.zeros(env.action_space.n))
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        episode = generate_episode(env)
        state, action, reward = zip(*episode)
        discounts = [gamma**i for i in range(len(reward)+1)]
        if mc_approx=='avg':
            for i, (s, a) in enumerate(zip(state, action)):
                returns_sum[s][a] += sum(r*d for r, d in zip(reward[i:], discounts[:-(1+i)]))
                N[s][a] += 1 
                Q[s][a] = returns_sum[s][a] / N[s][a]
        elif mc_approx == 'first':
            seen = set()
            for i, (s, a) in enumerate(zip(state, action)):
                if s not in seen: 
                    returns_sum[s][a] += sum(r*d for r, d in zip(reward[i:], discounts[:-(1+i)]))
                    N[s][a] += 1 
                    Q[s][a] = returns_sum[s][a] / N[s][a]
                    seen.add(s)
        else:
            raise ValueError(f"Unknown MC visitor approximation method {mc_approx}. Can be 'first' or 'avg'")
            
    return Q
````

### Part 2: MC Control

In this section, you will write your own implementation of constant-$\alpha$ MC control.  

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.
- `policy`: This is a dictionary where `policy[s]` returns the action that the agent chooses after observing state `s`.

(_Feel free to define additional functions to help you to organize your code._)

````python
def mc_control(env, num_episodes, alpha, gamma=1.0):
    """ Simple MC control algorithm.
    Args:
        env: This is an instance of an OpenAI Gym environment.
        num_episodes: This is the number of episodes that are 
            generated through agent-environment interaction.
        alpha: This is the step-size parameter for the update step.
        gamma: This is the discount rate. It must be a value 
            between 0 and 1, inclusive (default value: 1).
    
    Output:
        tuple: (Q, policy) where Q is a dictionary (of one-
            dimensional arrays) where Q[s][a] is the estimated
            action value corresponding to state s and action a 
            and policy is a dictionary where policy[s] returns 
            the action that the agent chooses after observing state s.
    """
    nA = env.action_space.n
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(nA))
    policy = defaultdict(lambda: np.argmax(np.random.uniform(0, 1, size=nA)))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
    
        episode = []
        state = env.reset()
        while True:
            action = policy[state]
            next_state, reward, done, info = env.step(action)
            episode.append((state, action, reward))
            state = next_state
            if done:
                break
        
        states, actions, rewards = zip(*episode)
        
        discounts = [gamma ** i for i in range(len(rewards) + 1)]
        # only way to hash pairs
        seen = list()
        for i, (s, a) in enumerate(zip(states, actions)):
            if [s, a] not in seen:
                G_t = sum([r * d for r, d in zip(rewards[i:], discounts[:-(i+1)])])
                Q[s][a] = Q[s][a] + alpha * (G_t - Q[s][a])
                policy[s] = np.argmax(Q[s])
 
    return policy, Q
````

## First visit Constant-$\alpha$ GLIE MC control

````python
def episode_generator_with_policy(env, policy, epsilon=0):
    """ Generates an episode with a given policy.
    The epsilon parameter is used to choose whether to
    follow the given policy or randomly choose an
    action given a state provided by the environment.
        Args:
            env: This is an instance of an OpenAI Gym environment.
            policy: This is the policy that gives an action
                for a provided state
            epsilon: This is the parameter that adds randomness to 
                the greedy algorithm. It must be a value between 0 and 1, 
                inclusive.
    """
    nA = env.action_space.n
    episode = []
    state = env.reset()
    while True:
        # epsilon greedy strategy for policy learning
        p = get_probs(policy[state], epsilon, nA)
        action = np.random.choice(np.arange(nA), p=p)
        next_state, reward, done, info = env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode


def get_probs(action, epsilon, nA):
    """ obtains the action probabilities corresponding to epsilon-greedy policy """
    policy_s = np.ones(nA) * epsilon / nA
    policy_s[action] = 1 - epsilon + (epsilon / nA)
    return policy_s

def mc_control_glie_constant_alpha(env, num_episodes, alpha,
                                   epsilon, epsilon_decay=0.99, 
                                   min_epsilon = 0.1, gamma=1.0):
    """ Extension of MC control algorithm.
        Adds epsilon learning, which adds randomness to the 
        greedy policy algorithm.
        Adds alpha parameter used to update the Q-value.
        
        Args:
            env: This is an instance of an OpenAI Gym environment.
            num_episodes: This is the number of episodes that are 
                generated through agent-environment interaction.
            alpha: This is the step-size parameter for the update step.
            epsilon: This is the parameter that adds randomness to 
                the greedy algorithm. It must be a value between 0 and 1, 
                inclusive.
            epsilon_decay: This is the decay of epsilon using exponential 
                decay. It must be a value between 0 and 1, inclusive 
                (default value: 0.99).
            min_epsilon: This is the lowest value epsilon can assume.
                It must be a value between 0 and 1, inclusive 
                (default value: 0.1).
            gamma: This is the discount rate. It must be a value 
                between 0 and 1, inclusive (default value: 1).

        Output:
            tuple: (Q, policy) where Q is a dictionary (of one-
                dimensional arrays) where Q[s][a] is the estimated
                action value corresponding to state s and action a 
                and policy is a dictionary where policy[s] returns 
                the action that the agent chooses after observing state s.
    """
    nA = env.action_space.n
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(nA))
    policy = defaultdict(lambda: np.argmax(np.random.uniform(0, 1, size=nA)))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # exponential decay with floor min_epsilon
        epsilon_i = max(min_epsilon, epsilon * np.exp(-epsilon_decay*i_episode))
        
        # generate episode
        episode = episode_generator_with_policy(env, policy, epsilon_i)
        
        states, actions, rewards = zip(*episode)
        
        discounts = [gamma ** i for i in range(len(rewards) + 1)]
        # only way to hash pairs to store seen states
        seen = list()
        for i, (s, a) in enumerate(zip(states, actions)):
            if [s, a] not in seen:
                G_t = sum([r * d for r, d in zip(rewards[i:], discounts[:-(i+1)])])
                Q[s][a] = Q[s][a] + alpha * (G_t - Q[s][a])
                policy[s] = np.argmax(Q[s])
 
    return policy, Q
````

````python
policy, Q = mc_control_glie_constant_alpha(env, 500000, alpha=0.02, epsilon=1.0, 
                                           epsilon_decay=0.99, min_epsilon=0.05)
V = dict((k,np.max(v)) for k, v in Q.items())
plot_blackjack_values(V)
````
